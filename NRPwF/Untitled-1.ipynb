```python
import pandas as pd
from yaml import load, dump, safe_load, safe_load_all, unsafe_load_all, safe_load
try:
    from yaml import CLoader as Loader, CDumper as Dumper
except ImportError:
    from yaml import Loader, Dumper
import os
from collections import defaultdict
import jmespath
from pathlib import Path
from ruamel.yaml import YAML
```

```python
fix_file = Path(os.path.join(os.getcwd(), 'results', '1_Gecode_BB_sol.yml'))
df = defaultdict(list)
# Parse the cleaned content
with fix_file.open('r') as f:
    content = next(f)
    for l in f:
        if l.startswith('- configuration:'):
            d = safe_load(content)
            df['configuration'].append(jmespath.search('[0].configuration', d))
            df['data_file'].append(jmespath.search('[0].data_file', d))
            df['model'].append(jmespath.search('[0].model', d))
            df['obj'].append(jmespath.search('[0].solution.obj', d))
            df['time'].append(jmespath.search('[0].time', d))
            content = l
            continue
        content += l

df = pd.DataFrame(df)
df
```
